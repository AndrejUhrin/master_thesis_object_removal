```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../human_validation_test_results/validation_customer_view.csv")
```

```python
df.head()
```

```python
df['Status'].value_counts()
```

```python
ax = (df['Status'].value_counts().plot.bar(figsize=(8, 6), rot=0, title='Pass vs. Fail results based on customer validation'))

ax.bar_label(ax.containers[0], label_type='edge')

ax.set(xlabel='Status', ylabel='Count')

plt.tight_layout()
plt.show()
```

```python
ax = (df['Status'].value_counts().plot.pie(figsize=(8, 6), colors=plt.cm.Pastel1.colors[:2][::-1], autopct=lambda pct: f"{pct:.1f}% ({round(pct * len(df) / 100):d})", startangle=90, ylabel=''))

ax.axis('equal')               
plt.title('Pass vs. Fail results based on customer validation')
plt.tight_layout()
plt.show()
```
